```python
import pandas as pd

# Load the two CSV files
csv_file1 = 'csv data/mpmb.csv'  # The file that contains the 'C/N0' column
csv_file2 = 'csv data/etsb.csv'  # The file where 'C/N0' column needs to be added

df1 = pd.read_csv(csv_file1)
df2 = pd.read_csv(csv_file2)

# Ensure the column names are correct by printing out the column names
print("Columns in dataset1:", df1.columns)
print("Columns in dataset2:", df2.columns)

df2 = df2.rename(columns={'PRN Number': 'PRN', 'Time of Week': 'Seconds of Week'})

# Merge the datasets based on 'PRN' and 'Seconds of Week' columns
merged_df = pd.merge(df1, df2[['PRN', 'Seconds of Week', 'C/N0']], on=['PRN', 'Seconds of Week'], how='left')


# Print the merged DataFrame to verify
print(merged_df.head())
print(merged_df.tail())
```
